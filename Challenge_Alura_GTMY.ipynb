## Recomendación de tienda a vender

En este informe se presenta un análisis de desempeño de las cuatro tiendas de la cadena **Alura Store**, con el objetivo de recomendar cuál de ellas debería ser vendida por el Sr. Juan para financiar un nuevo emprendimiento.  
El análisis se basó en datos reales de ventas, productos, reseñas de clientes y costos de envío.  
Se utilizaron herramientas de ciencia de datos con Python, principalmente `pandas` y `matplotlib`, para extraer información valiosa y presentar conclusiones basadas en evidencia.

---

##  Desarrollo del Análisis

### 1. Ingresos Totales por Tienda

Los ingresos totales calculados fueron:

- **Tienda 1**: $1,150,880,400  
- **Tienda 2**: $1,116,343,500  
- **Tienda 3**: $1,098,019,600  
- **Tienda 4**: $1,038,375,700  

 **Conclusión**: La Tienda 4 tiene la facturación más baja de las cuatro.

---

### 2. Categorías de Productos Más Vendidas

Se identificaron los 10 productos más vendidos y los 10 menos vendidos en cada tienda. Estos datos permiten evaluar qué tan bien rota el inventario y qué tan efectiva es la tienda en impulsar productos de alto movimiento.

**Ejemplos destacados:**

- **Tienda 1**: *Microondas* y *TV LED UHD 4K* vendieron 60 unidades; *Celular ABXY* y *Auriculares con micrófono* solo 33.  
- **Tienda 2**: *Iniciando en programación* lideró con 65 ventas; *Juego de mesa* e *Impresora* apenas llegaron a 32–34 ventas.  
- **Tienda 3**: *Kit de bancas* vendió 57 unidades; productos como *Bloques de construcción*, *Mochila* y *Microondas* vendieron 35–36.  
- **Tienda 4**: *Cama box* fue el más vendido con 62 unidades; *Guitarra eléctrica* y *Armario* no superaron las 33–34 ventas.

 La **Tienda 4** tiene una mayor proporción de productos con bajo rendimiento, lo cual impacta negativamente su eficiencia general.

---

### 3. Calificaciones Promedio de Clientes

- **Tienda 1**: 3.976  
- **Tienda 2**: 4.037  
- **Tienda 3**: 4.048  
- **Tienda 4**: 3.995  

 **Conclusión**: Las tiendas 1 y 4 tienen las calificaciones más bajas por parte de los clientes.

---

### 4. Productos más y menos vendidos

La evaluación por tienda mostró que, aunque todas presentan productos con poca rotación, **la Tienda 4 aparece más veces entre los productos menos vendidos**.

- Tienda 4 tiene artículos como *Guitarra eléctrica* y *Armario* con tan solo 33–34 unidades vendidas.  
- En contraste, otras tiendas logran mejores cifras en productos similares.

 Esto refleja baja demanda y escasa rotación, afectando su rentabilidad.

---

### 5. Costo Promedio de Envío

Los costos promedios de envío fueron:

- **Tienda 1**: 26,018.61  
- **Tienda 2**: 25,216.24  
- **Tienda 3**: 24,805.68  
- **Tienda 4**: 23,459.46  

 Aunque **Tienda 4** tiene el costo de envío más bajo, esta ventaja logística **no compensa** su bajo nivel de facturación, su menor diversidad de productos vendidos y la baja rotación de inventario.

---

##  Conclusión

Luego de analizar detalladamente los datos de las cuatro tiendas de **Alura Store**, se recomienda que el Sr. Juan **venda la Tienda 4**. Esta decisión se sustenta en los siguientes puntos clave:

- Es la tienda con la **facturación más baja**: $1,038,375,700.
- Presenta **menor diversidad de productos** vendidos por categoría.
- Tiene una de las **calificaciones de cliente más bajas** (3.995).
- Posee **más productos con baja rotación**, afectando ingresos potenciales.
- Aunque tiene el **menor costo de envío promedio**, no logra destacar en otras métricas clave.

 En conjunto, estos factores muestran que la **Tienda 4 es la menos eficiente y rentable**.  
Venderla permitiría al Sr. Juan liberar capital e invertirlo en un nuevo negocio con mayor potencial.
